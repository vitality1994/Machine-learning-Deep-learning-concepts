```python
from sklearn.datasets import load_diabetes
diabetes = load_diabetes()  #for convinience.
```

```python
print(diabetes.data.shape, diabetes.target.shape)
# Shape method will show the shape of a matrix.
# In the input data set, there are 442 samples.
# Each sample has 10 features (= independent variables = Explanatory variables)
# Target data has the same amount of samples as input data.
```

```python
print(diabetes.data[0:3]) # Feature values of the first three samples.
print(diabetes.target[0:3]) # Target values of the first three samples.

# We do not know what weights are assigned to each feature, but with some specific weights
```

```python
print(diabetes.data[0], diabetes.target[0])
# First array of input data corresponds with first target data.
# n th array of input data corresponds with n th target data.
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize = (16, 10))
plt.subplot(331)
plt.scatter(diabetes.data[:, 0], diabetes.target)
plt.subplot(332)
plt.scatter(diabetes.data[:, 1], diabetes.target) 
plt.subplot(333)
plt.scatter(diabetes.data[:, 2], diabetes.target) 
plt.subplot(334)
plt.scatter(diabetes.data[:, 3], diabetes.target) 
plt.subplot(335)
plt.scatter(diabetes.data[:, 4], diabetes.target)
plt.subplot(336)
plt.scatter(diabetes.data[:, 5], diabetes.target) 
plt.subplot(337)
plt.scatter(diabetes.data[:, 6], diabetes.target)
plt.subplot(338)
plt.scatter(diabetes.data[:, 7], diabetes.target) 
plt.subplot(339)
plt.scatter(diabetes.data[:, 8], diabetes.target) 
plt.xlabel('x')
plt.ylabel('y')
plt.show()

# from 1st to 9th features, 
# We can visibly check, 3rd, 7th, 9th features have a linear relationship with target data.
```
