```python
from IPython.display import Image


# In this chapter, we will see how we can minimize the error between target data and predicted value.
# If we want to express 'Gradient descent' more technically, we can say 
# 'minimize the value of a loss function(the difference between target data and predicted value)'.

# Now, we will see how 'squared error', which is one of the loss function, will work.
```

```python
from sklearn.datasets import load_diabetes
diabetes = load_diabetes()

x = diabetes.data[:, 2]
y = diabetes.target

w = 1.0
b = 1.0

for i in range(1, 100):
    for x_i, y_i in zip(x, y):
        y_hat = x_i * w + b
        err = y_i - y_hat
        w_rate = x_i
        w = w + w_rate * err
        b = b + 1 * err
        
import matplotlib.pyplot as plt
plt.scatter(x, y)
pt1 = (-0.1, -0.1 * w + b)
pt2 = (0.15, 0.15 * w + b)
plt.plot([pt1[0], pt2[0]], [pt1[1], pt2[1]])
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
